```python
# Step 1: Import Libraries

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Step 2: Load Data

url = "https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv"
df = pd.read_csv(url)
print(df.head())
```

```python
# Step 3: Visualize the Data

plt.scatter(df['Hours'],df['Scores'])
plt.title('Hours vs Percentage')
plt.xlabel('Hours Studied')
plt.ylabel('Score')
plt.grid('True')
plt.show() # for other IDE 
```

```python
X = df[['Hours']]
y = df['Scores']
print(X)
```

```python
# Step 5: Train/Test Split (80/20)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Step 6: Create and Train the Model
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Step 7: Predict
y_pred = model.predict(X_test)
```

```python
# Step 8: Compare Actual vs Predicted
df_result = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
print(df_result)
```

```python
# Step 9: Visualize Prediction
plt.scatter(X_test, y_test, color='red', label='Actual')
plt.plot(X_test, y_pred, color='blue', linewidth=2, label='Predicted Line')
plt.title('Actual vs Predicted')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.legend()
plt.grid(True)
```

```python
# Step 10: Evaluate Model
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
print("R2 Score (Accuracy):", r2_score(y_test, y_pred))
```

```python
hours = 9.25
predicted_score = model.predict([[hours]])
print(f"If a student studies for {hours} hours, they may score {predicted_score[0]:.2f} marks.")
```
